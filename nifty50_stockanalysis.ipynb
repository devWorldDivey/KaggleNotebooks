```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #data Visualization
import seaborn as sns #data Visualization
import plotly.io as pio
```

```python
df=pd.read_csv("../input/nifty50-stocks-dataset/National_Stock_Exchange_of_India_Ltd.csv")
df
```

```python
df.dtypes
```

```python
df['Open'] = df['Open'].str.replace(',','')
df['Open'] = df['Open'].astype(float)

df['High'] = df['High'].str.replace(',','')
df['High'] = df['High'].astype(float)

df['Low'] = df['Low'].str.replace(',','')
df['Low'] = df['Low'].astype(float)

df['LTP'] = df['LTP'].str.replace(',','')
df['LTP'] = df['LTP'].astype(float)

df['Turnover (crs.)'] = df['Turnover (crs.)'].str.replace(',','')
df['Turnover (crs.)'] = df['Turnover (crs.)'].astype(float)

df['52w H'] = df['52w H'].str.replace(',','')
df['52w H'] = df['52w H'].astype(float)

df['52w L'] = df['52w L'].str.replace(',','')
df['52w L'] = df['52w L'].astype(float)
```

```python
for cols in str_cols:
    print("Column : ", cols)
    print(f"Value counts of {cols}: ",df[cols].value_counts().sum())
    print(f"Number of Unique Values in {cols}: ",df[cols].nunique())
    print()
```

```python
bank_df = df['Symbol'].isin(['AXISBANK','HDFCBANK','ICICIBANK','INDUSINDBK','KOTAKBANK','SBIN','HDFC','BAJAJFINSV','BAJFINANCE','HDFCLIFE'])
bank_table = df[bank_df].pivot_table(values =['Open','High','Low','LTP','Chng','Volume (lacs)','52w H','52w L','Turnover (crs.)','365 d % chng','30 d % chng'], index= ['Symbol'])
bank_table.reset_index( inplace=True)
bank_table
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)

sns.barplot(data = bank_table, x = 'Symbol', y = '365 d % chng', palette='Blues_d')

plt.xlabel("Symbol")  
plt.ylabel('% Change in a Year')  
plt.title("Yearly Growth")  

plt.show
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.barplot(data = bank_table, x = 'Symbol', y = '30 d % chng',palette='mako')
```

```python
sort_yearly=df.sort_values('365 d % chng')
sort_yearly
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=90)


sns.lineplot(data = sort_yearly, x = 'Symbol', y = '365 d % chng')
plt.xlabel("Symbol")  
plt.ylabel('% Change in a Year(Sorted)')  
plt.title("Yearly Growth(Sorted)")  
plt.show
```
